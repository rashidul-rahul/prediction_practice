```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
data = pd.read_csv("canada_per_capita_income.csv")
data.head(3)
```

```python
%matplotlib inline
plt.xlabel("Year", fontsize="15")
plt.ylabel("Per Capita Income (US$)", fontsize="15")
plt.scatter(data.year, data["per capita income (US$)"], marker="+", color="green")
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[["year"]], data["per capita income (US$)"])
reg.predict([[2023]])
```

```python
w_data = pd.read_csv("years.csv")
net_income = reg.predict(w_data[["year"]])
w_data["per capita income (US$)"] = net_income
w_data.to_csv("prediction.csv", index=False)
```

```python
%matplotlib inline
plt.xlabel("Year", fontsize=13)
plt.ylabel("per capita income (US$)", fontsize=13)
plt.scatter(data.year, data["per capita income (US$)"], marker="+", color="green")
plt.plot(data.year, reg.predict(data[["year"]]), color="red")
```
